## Dataset

Global retail sales dataset containing 51,290 transactions across different markets, regions, customers, products, and categories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option("display.max_columns", None)
```

The dataset contains 51,290 sales records and 21 columns including order details, customer information, product details, sales, profit, and shipping information.

```python
df["sales"] = (
    df["sales"]
    .astype(str)
    .str.replace(",", "")
    .astype(float)
)
```

```python
df["order_date"] = pd.to_datetime(
    df["order_date"],
    format="mixed",
    dayfirst=True
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    format="mixed",
    dayfirst=True
)
```

```python
df["shipping_days"] = (
    df["ship_date"] - df["order_date"]
).dt.days
```

```python
df["shipping_days"].describe()
```

Shipping days was created to measure delivery time and allow future analysis of logistics performance.

```python
total_sales = df["sales"].sum()
total_profit = df["profit"].sum()

profit_margin = (
    total_profit / total_sales
) * 100

total_orders = df["order_id"].nunique()

total_customers = df["customer_name"].nunique()

average_order_value = (
    total_sales / total_orders
)
```

```python
print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
print("Profit Margin:", profit_margin)
print("Orders:", total_orders)
print("Customers:", total_customers)
print("Average Order Value:", average_order_value)
```

# Market Performance Analysis

```python
market_analysis = df.groupby("market").agg(
    total_sales=("sales", "sum"),
    total_profit=("profit", "sum"),
    orders=("order_id", "nunique")
)

market_analysis
```

```python
market_analysis.sort_values(
    by="total_sales",
    ascending=False
)
```

```python
market_analysis["profit_margin"] = (
    market_analysis["total_profit"] /
    market_analysis["total_sales"]
) * 100

market_analysis.sort_values(
    by="profit_margin",
    ascending=False
)
```

# Category Profitability Analysis

```python
category_analysis = df.groupby("category").agg(
    total_sales=("sales", "sum"),
    total_profit=("profit", "sum"),
    orders=("order_id", "nunique")
)

category_analysis["profit_margin"] = (
    category_analysis["total_profit"] /
    category_analysis["total_sales"]
) * 100

category_analysis.sort_values(
    by="total_profit",
    ascending=False
)
```

# Product Performance Analysis

```python
product_analysis = df.groupby("product_name").agg(
    total_sales=("sales", "sum"),
    total_profit=("profit", "sum"),
    quantity=("quantity", "sum")
)

product_analysis.sort_values(
    by="total_sales",
    ascending=False
).head(10)
```

# Discount Impact Analysis

```python
discount_analysis = df.groupby("discount").agg(
    total_sales=("sales", "sum"),
    total_profit=("profit", "sum"),
    orders=("order_id", "nunique")
)

discount_analysis
```

```python
loss_analysis = df[df["profit"] < 0].groupby("discount").agg(
    loss_sales=("sales", "sum"),
    total_loss=("profit", "sum"),
    orders=("order_id", "nunique")
)

loss_analysis.sort_values(
    by="total_loss"
)
```

# Final Business Recommendations

1. Reduce excessive discounts because discounts above 40% significantly reduce profitability.

2. Increase focus on Technology products due to strong profit margins.

3. Investigate Furniture profitability because sales are high but margins are lower.

4. Analyze EMEA performance because it has the weakest profit margin.

5. Promote high-profit products instead of only high-volume products.
